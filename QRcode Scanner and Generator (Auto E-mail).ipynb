# ...final QRcode Software.

### In this project, it will Generate QRcode from Text, then it will mail it my friends and bosses. Also, it can Scan QRcode via laptop webcam, or mobile camera...

## Code in right side is Multi-Clipboard, which can hold multiple Copied text.

```python
import cv2, time, os
from pyzbar.pyzbar import decode
from PIL import Image
import smtplib, getpass, os
from email.mime.multipart import MIMEMultipart 
from email.mime.text import MIMEText 
from email.mime.base import MIMEBase 
from email import encoders 
import qrcode, os
from PIL import Image

while True: # ...enter string input to EXIT loop.
    
    print('\n-------------------------------------------------\n')
    print('1). QRcode Generator')
    print('2). QRcode Scanner')
    print('\n...else break')

    try:
        choice = int(input('\nEnter your choice (int) : '))
    except:
        input('Press Enter to EXIT...')
        break
        
    if choice == 1:
        
        print('\n===============================================\n')
        
        # data = 'https://www.facebook.com/Imvickykumar999/' 
                                    # ...this is link of my fb account.
        data = input('Paste TEXT to convert into QRCODE : ')
        
#         qr = qrcode.make(data)
#         photo = r'C:\Users\Vicky Kumar\Pictures\Saved Pictures\qr.jpg'
#         qr.save(photo)
#         os.startfile(photo)
#         qr

        img = qrcode.QRCode(
            version=1,
            error_correction=qrcode.constants.ERROR_CORRECT_L,
            box_size=10,
            border=4,
        )

        img.add_data(data)
        img.make(fit=True)
        
        print('\nOption of COLORS are : red, white, black, green, yellow, pink, blue, ...etc.\n')
        color_set = ['red', 'white', 'black', 'green', 'yellow', 'pink', 'blue', '']
        
        print('\n * is compulsory fields...( ~ write anything !!!)\n')
        fill = input('Enter *Fill Color of QRcode : ')
        back = input('Enter *Background Color of QRcode : ')
        
        if fill and back not in color_set:
            fill = 'black'
            back = 'white'
            print('\nSince, there was SPELLING mistake, so DEFAULT color, {} is used.. !!!\n'.format((fill, back)))
            
        photo = r'C:\Users\Vicky Kumar\Pictures\Saved Pictures\qr.jpg'
        img = img.make_image(fill_color = fill, back_color = back)
        img.save(photo)
        
        print('\nImage file of QRcode is Opened...\n')
        os.startfile(photo)
#         img

        fromaddr = "imvickykumar999@gmail.com"

        toaddr = ["imvickykumar999@gmail.com",
                  "ankitmalpani177@gmail.com",
                  "sv6375261073@gmail.com",
                  "stuti24m@gmail.com",
                  "aryasingh.rathore123@gmail.com"]
        
        def show(toaddr):
            c = 0
            print()
            for t in toaddr:
                print(c+1, '). ', t)
                c = c + 1
            
        show(toaddr)
        
        email_id = input('\nEnter NEW E-mail ID to (Append) : ')
        
        def Append(email_id):
            toaddr.append(email_id)
            show(toaddr)
            
        Append(email_id)
        
        try:
            n = int(input('\nEnter index Whom to Mail.. ? (int) : ')) -1
            print('\nYou have chosen to mail to : {}'.format(toaddr[n]))
        except Exception as x:
            print('\nError : {}'.format(x))
            continue

        # instance of MIMEMultipart 
        msg = MIMEMultipart() 

        msg['From'] = fromaddr 
        msg['To'] = toaddr[n]
        msg['Subject'] = "Auto mailing QRcode : " + input('\nEnter [ SUBJECT ] of E-mail : ')

        body = '''
        ...this email is sent using python code !!!
        ==========================================

        ...an attachment of QRcode is mailed to you : {}

        '''.format(data)

        msg.attach(MIMEText(body, 'plain')) 
        filename = "qr.jpg"
        path = os.path.join('C:\\Users', os.getlogin(), 'Pictures', 'Saved Pictures', filename)
        attachment = open(path, "rb") 

        # instance of MIMEBase and named as p 
        p = MIMEBase('application', 'octet-stream') 

        # To change the payload into encoded form 
        p.set_payload((attachment).read()) 

        encoders.encode_base64(p) 
        p.add_header('Content-Disposition', "attachment; filename= %s" % filename) 
        msg.attach(p) 

        # creates SMTP session 
        s = smtplib.SMTP('smtp.gmail.com', 587) 

        # start TLS for security 
        s.starttls() 
        
        def passw():
            return getpass.getpass('Enter Your Gmail Password : ')
            
        try:
            password = passw()
            s.login(fromaddr, password)
            
        except Exception as e:
            print('\nError : {}'.format(e))
            print('\n>>> Wrong Password Entered... >>> [ Last Chance ], TRY AGAIN !!!\n')
            password = passw()
            try:
                s.login(fromaddr, password)
            except:
                print('\n>>> !!! AGAIN, Wrong Password Entered...\n')
                if input('\nWanna see Password.. ? (y/n) : ') == 'y':
                    print(password)
                continue
                
#         finally:
#             print()
            
        text = msg.as_string()
        s.sendmail(fromaddr, toaddr[n], text)
        s.quit() 

        print('\n...Email is sent Successfully to {}'.format(toaddr[n]))

    # ========================================================================

    elif choice == 2:

        print('\n..which camera wanna use ?\n')
        print('1). Laptop')
        print('2). IP WebCam')
        print('\n...any-thing else to EXIT.')
        camera = input('\nWhich Camera wanna use.. ? (if 2, IP WebCam must be ON...) : ')

        if camera == '1':
            url = 0
        elif camera == '2':
            url = 'http://10.71.220.102:8080/video'
        else:
            break

        video = cv2.VideoCapture(url)

        a = 1
        photo = r'C:\Users\Vicky Kumar\Pictures\Saved Pictures\qr.jpg'

        while True:

            a = a + 1
            check, frame = video.read()
        #     print(frame)

        #     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
            cv2.putText(frame, "PRESS q to Click...", (60, 30) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.imshow('Capturing', frame)
            key = cv2.waitKey(30)
        #     print(key)
            if key == ord('q'):
                break

        # print(a)
        # print(ord('q'))
        video.release()
        cv2.destroyAllWindows()

        cv2.imwrite(photo, frame)
#         cv2.imshow('Photo : ' + str(check), frame)
#         cv2.waitKey(5000)
        os.startfile(photo)

        try:
            d = decode(Image.open(photo))
            print('\n\t>>> Text in QRcode : {}'.format(d[0].data.decode()))
        except Exception as e:
            print('\n>>>> Error (no QRcode found...) : {}'.format(e))

    else:
        break
        
print('\n...exited,\n\t>>> Thanks for using my QRcode Code !!!')
```
